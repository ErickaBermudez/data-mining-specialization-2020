```python
import pandas as pd
```

```python
data = pd.read_csv("data-1.csv");
data
```

```python
data = data[data['year'] == 2014]
data
```

```python
countries_2014 = data.groupby('country').aggregate('sum')
countries_2014 = countries_2014[['suicides/100k pop']]
countries_2014 
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
# Control del tamaÃ±o de la figura del mapa
fig, ax = plt.subplots(figsize=(10, 10))
 
# Control del tÃ­tulo y los ejes
ax.set_title('Suicides per 100k population in 2014', 
             pad = 20, 
             fontdict={'fontsize':20, 'color': '#4873ab'})
ax.set_xlabel('Longitud')
ax.set_ylabel('Latitud')
 
# Mostrar el mapa finalizado
countries_2014.plot(column='suicides/100k pop', cmap='plasma', ax=ax, zorder=5)
```
