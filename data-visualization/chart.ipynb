```python
import pandas as pd
```

```python
data = pd.read_csv("data-1.csv");
data
```

```python
data = data[data['year'] == 2014]
data
```

```python
countries_2014 = data.groupby('country').aggregate('sum')
countries_2014 = countries_2014[['suicides/100k pop']]
countries_2014
```

```python
countries_2014
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
world
```

```python
map_data = world.merge(countries_2014, how="left", left_on = 'name', right_on = 'country')
map_data
```

```python
map_data = map_data[(map_data.continent == 'South America')]
map_data
```

```python
ax = map_data.plot(column='suicides/100k pop', 
                            missing_kwds = dict(color = "lightgrey", hatch = "///", label="Missing Values"),
                            cmap = 'YlGnBu', figsize=(15,9), 
                            k=5, legend =  True,
                           );
ax.set_title('Suicides per 100k of population in 2014 in South America', fontdict= 
            {'fontsize':20})

ax.set_axis_off()
```
